```python
import pandas as pd
import os
import bar_chart_race as bcr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_future_population')
df
```

```python
df5080 = df[0]
df5080
```

```python
df8515 = df[1]
df8515.reset_index(drop=True, inplace=True)
df8515
```

```python
outer_merged = pd.merge(df5080, df8515, how="inner", on='Country (or dependent territory)')
merged_df = outer_merged.drop(['%_x',
       '%.1_x', '%.2_x', '%.3_x', '%.4_x',
       '%.5_x', '%_y', '%.1_y', '%.2_y',
       '%.3_y', '%.4_y', '%.5_y', '%.6'], axis=1)
merged_df.reset_index(drop=True, inplace= True)
sorted_df = merged_df.sort_values(by="1950",ascending=False,kind="mergesort")
finaldf = sorted_df.transpose()
file_name = 'populationworkbook.csv'
finaldf.to_csv(file_name)
```

```python
merged_df.sort_values(by="1950",ascending=False,kind="mergesort")
```

```python
df3 = pd.read_csv(file_name)
```

```python
df4 = df3.iloc[: ,:]
df4
```

```python
df4.columns = df4.loc[0,:]
df4.drop(0,axis=0,inplace=True)
df4
```

```python
df4.rename(columns={'Country (or dependent territory)':'Date'}, inplace=True)
df4
```

```python
df4.set_index('Date', inplace=True)
```

```python
df5 = pd.read_csv('worldPopulation.csv')
```

```python
def summary(values, ranks):
    population = int(round(values.sum(), -2))
    s = f'TOTAL POPULATION - {population:,.0f}'
    return {'x': .99, 'y': .05, 's': s, 'ha': 'right', 'size': 8}
```

```python
bcr.bar_chart_race(df=df5, n_bars=27, filter_column_colors=True, bar_size=0.6, 
                   title='Population by Country per Year',
                   title_size='small', fixed_max=True, bar_label_size=4,
                   steps_per_period=9, period_length=420, perpendicular_bar_func='mean', period_summary_func=summary)
```

```python
df6 = df5.copy()
df6.drop('World', axis=1, inplace=True)
df6['Date'] = pd.to_datetime(df6['Date'].astype(str), format='%Y')
df6.set_index('Date',inplace=True)
df6
```

```python
bcr.bar_chart_race(df=df6, n_bars=27, filter_column_colors=True, bar_size=0.6, 
                   title='Population by Country per Year', interpolate_period=True,
                   title_size='small', fixed_max=True, bar_label_size=4,
                  steps_per_period=9, period_length=420, perpendicular_bar_func='mean', period_summary_func=summary)
```
